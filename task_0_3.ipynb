```python
# Install required libraries (only needed in Colab)
!pip install requests beautifulsoup4

# Import libraries
import requests
from bs4 import BeautifulSoup

# URL of the news website (Example: BBC News)
url = "https://www.bbc.com/news"

# Send HTTP request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    print("✅ Successfully fetched the webpage")
else:
    print("❌ Failed to fetch webpage")
    exit()

# Parse HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find headline tags (BBC uses h2 and h3 sometimes)
headlines = soup.find_all(['h2', 'h3'])

# Extract text from tags
headline_list = []

for tag in headlines:
    text = tag.get_text(strip=True)

    # Filter out empty or very short text
    if text and len(text) > 20:
        headline_list.append(text)

# Remove duplicates
headline_list = list(set(headline_list))

# Display headlines
print("\n📰 Top Headlines:\n")
for i, headline in enumerate(headline_list, 1):
    print(f"{i}. {headline}")

# Save headlines to a text file
file_name = "news_headlines.txt"

with open(file_name, "w", encoding="utf-8") as file:
    for headline in headline_list:
        file.write(headline + "\n")

print(f"\n📁 Headlines saved to {file_name}")
```
